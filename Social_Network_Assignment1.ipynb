Part 4

```python
#for undirected graph
# degree centrality
degree_centrality = nx.degree_centrality(G_undirected)
print("Degree Centrality:")
print(degree_centrality)

# closeness centrality
closeness_centrality = nx.closeness_centrality(G_undirected)
print("\nCloseness Centrality:")
print(closeness_centrality)

# betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G_undirected)
print("\nBetweenness Centrality:")
print(betweenness_centrality)

# eigenvector centrality
eigenvector_centrality = nx.eigenvector_centrality(G_undirected)
print("\nEigenvector Centrality:")
print(eigenvector_centrality)

# calculate PageRank
pagerank = nx.pagerank(G_undirected)
print("PageRank:")
print(pagerank)
```

```python
def degrees_round(s_t_deg):
    s_t_deg = dict(sorted(s_t_deg.items()))
    for key in s_t_deg:
        s_t_deg[key] = round(s_t_deg[key], 2)
    return s_t_deg
```

```python
udg_deg_centrality = degrees_round(degree_centrality)
udg_closeness_centrality = degrees_round(closeness_centrality)
udg_betweeness_centrality = degrees_round(betweenness_centrality)
udg_eigen_vector_centrality = degrees_round(eigenvector_centrality)
udg_page_rank = degrees_round(pagerank)

df = pd.DataFrame.from_dict(udg_deg_centrality,orient='index',columns=['Degree_Centrality'])
df = df.assign(Closeness_Centrality= udg_closeness_centrality.values())
df = df.assign(Betweeness_Centrality= udg_betweeness_centrality.values())
df = df.assign(Eigen_Vector_Centrality= udg_eigen_vector_centrality.values())
df = df.assign(Page_Rank = udg_page_rank.values())
```

```python
# find minimum and maximum centrality scores and their corresponding nodes
min_degree_node = min(degree_centrality,key=degree_centrality.get)
max_degree_node = max(degree_centrality,key=degree_centrality.get)
min_closeness_node = min(closeness_centrality, key=closeness_centrality.get)
max_closeness_node = max(closeness_centrality, key=closeness_centrality.get)
min_betweenness_node = min(betweenness_centrality, key=betweenness_centrality.get)
max_betweenness_node = max(betweenness_centrality, key=betweenness_centrality.get)
min_pagerank_node = min(pagerank, key=pagerank.get)
max_pagerank_node = max(pagerank, key=pagerank.get)
min_eigenvector_node = min(eigenvector_centrality, key=eigenvector_centrality.get)
max_eigenvector_node = max(eigenvector_centrality, key=eigenvector_centrality.get)

min_degree_score = round(degree_centrality[min_degree_node],2)
max_degree_score = round(degree_centrality[max_degree_node],2)
min_closeness_score = round(closeness_centrality[min_closeness_node],2)
max_closeness_score = round(closeness_centrality[max_closeness_node],2)
min_betweenness_score = round(betweenness_centrality[min_betweenness_node],2)
max_betweenness_score = round(betweenness_centrality[max_betweenness_node],2)
min_pagerank_score = round(pagerank[min_pagerank_node],2)
max_pagerank_score = round(pagerank[max_pagerank_node],2)
min_eigenvector_score = round(eigenvector_centrality[min_eigenvector_node],2)
max_eigenvector_score = round(eigenvector_centrality[max_eigenvector_node],2)

# print results
print("Degree Centrality:")
print("Minimum score:", min_degree_score, "for node", min_degree_node)
print("Maximum score:", max_degree_score, "for node", max_degree_node)

print("\nCloseness Centrality:")
print("Minimum score:", min_closeness_score, "for node", min_closeness_node)
print("Maximum score:", max_closeness_score, "for node", max_closeness_node)

print("\nBetweenness Centrality:")
print("Minimum score:", min_betweenness_score, "for node", min_betweenness_node)
print("Maximum score:", max_betweenness_score, "for node", max_betweenness_node)

print("\nEigen Vector Centrality:")
print("Minimum score:", min_eigenvector_score, "for node", min_eigenvector_node)
print("Maximum score:", max_eigenvector_score, "for node", max_eigenvector_node)

print("\nPageRank:")
print("Minimum score:", min_pagerank_score, "for node", min_pagerank_node)
print("Maximum score:", max_pagerank_score, "for node", max_pagerank_node)
```
